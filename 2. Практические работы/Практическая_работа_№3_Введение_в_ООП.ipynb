## **Задание №1. Создайте класс Sphere для представления сферы в трехмерном пространстве**


Обеспечьте следующие методы класса:

1. Конструктор, принимающий 4 действительных числа: радиус, и 3 координаты центра шара. Если конструктор вызывается без аргументов, создать объект сферы с единичным радиусом и центром в начале координат. Если конструктор вызывается с аргументом, создать объект сферы с соответствующим радиусом и центром в начале координат.

2. Метод get_volume (), который возвращает действительное число — объем шара, ограниченной текущей сферой.

3. Метод get_square (), который возвращает действительное число — площадь внешней поверхности сферы.

4. Метод get_radius (), который возвращает действительное число — радиус сферы.

5. Метод get_center (), который возвращает тьюпл с 3 действительными числами — координатами центра сферы в том же порядке, в каком они задаются в конструкторе.

6. Метод set_radius (r), который принимает 1 аргумент — действительное число, и меняет радиус текущей сферы, ничего не возвращая.

7. Метод set_center (x, y, z), который принимает 3 аргумента — действительных числа, и меняет координаты центра сферы, ничего не возвращая. Координаты задаются в том же порядке, что и в конструкторе.

8. Метод is_point_inside (x, y, z), который принимает 3 аргумента — действительных числа — координаты некоторой точки в пространстве (в том же порядке, что и в конструкторе), и возвращает логическое значение True или False в зависимости от того, находится эта точка внутри сферы.

````python
# Тесты

s0 = Sphere(0.5) # test sphere creation with radius and default center
print(s0.get_center()) # (0.0, 0.0, 0.0)
print(s0.get_volume()) # 0.523598775598
print(s0.is_point_inside(0 , -1.5, 0)) # False
s0.set_radius(1.6)
print(s0.is_point_inside(0, -1.5, 0)) # True
print(s0.get_radius()) # 1.6
````

````python
import math

class Sphere:
    """
    Класс для представления сферы в трёхмерном пространстве.
    """

    def __init__(self, radius: float = 1.0, x: float = 0.0, y: float = 0.0, z: float = 0.0):
        """
        Конструктор класса Sphere.
        Если вызывается без аргументов — создаёт сферу с радиусом 1 и центром в начале координат.
        Если передан только радиус — создаёт сферу с данным радиусом и центром (0, 0, 0).
        Если переданы все 4 аргумента — создаёт сферу с заданными параметрами.
        """
        self.radius = radius
        self.center = (x, y, z)

    def get_volume(self) -> float:
        """
        Возвращает объём сферы.
        Формула: V = 4/3 * π * r³
        """
        return (4 / 3) * math.pi * self.radius ** 3

    def get_square(self) -> float:
        """
        Возвращает площадь поверхности сферы.
        Формула: S = 4 * π * r²
        """
        return 4 * math.pi * self.radius ** 2

    def get_radius(self) -> float:
        """
        Возвращает радиус сферы.
        """
        return self.radius

    def get_center(self) -> tuple[float, float, float]:
        """
        Возвращает координаты центра сферы в виде кортежа (x, y, z).
        """
        return self.center

    def set_radius(self, r: float) -> None:
        """
        Устанавливает новый радиус сферы.
        """
        self.radius = r

    def set_center(self, x: float, y: float, z: float) -> None:
        """
        Устанавливает новые координаты центра сферы.
        """
        self.center = (x, y, z)

    def is_point_inside(self, x: float, y: float, z: float) -> bool:
        """
        Проверяет, находится ли точка (x, y, z) внутри сферы.
        Возвращает True, если точка внутри или на поверхности, иначе False.
        """
        cx, cy, cz = self.center
        distance = math.sqrt((x - cx) ** 2 + (y - cy) ** 2 + (z - cz) ** 2)
        return distance <= self.radius
s0 = Sphere(0.5) # test sphere creation with radius and default center
print(s0.get_center()) # (0.0, 0.0, 0.0)
print(s0.get_volume()) # 0.523598775598
print(s0.is_point_inside(0 , -1.5, 0)) # False
s0.set_radius(1.6)
print(s0.is_point_inside(0, -1.5, 0)) # True
print(s0.get_radius()) # 1.6
````

## **Задание №2. Создайте класс SuperStr, который наследует стандартный класс str и содержит 2 новых метода:**

1. Метод is_repeatance (s), который принимает 1 аргумент s и возвращает True или False в зависимости от того, может ли текущая строку быть получена целым количеством повторов строки s. Вернуть False, если s не является строкой. Считать, что пустая строка не содержит повторов.

2. Метод is_palindrom (), который возвращает True или False в зависимости от того, является ли строка [палиндромом](https://ru.wikipedia.org/wiki/Палиндром). Регистрами символов пренебрегать. Пустую строку считать палиндромом.

````python
class SuperStr(str):
    """
    Класс SuperStr — расширенная версия стандартного str.
    Добавляет методы:
      - is_repeatance(s)
      - is_palindrom()
    """

    def is_repeatance(self, s) -> bool:
        """
        Проверяет, можно ли получить текущую строку повторением строки s.
        Возвращает True или False.
        """
        # Проверка, что аргумент — строка
        if not isinstance(s, str) or s == "":
            return False

        # Длина текущей строки должна быть кратна длине s
        if len(self) % len(s) != 0:
            return False

        # Проверяем, можно ли составить self повторением s
        repeat_count = len(self) // len(s)
        return s * repeat_count == self

    def is_palindrom(self) -> bool:
        """
        Проверяет, является ли текущая строка палиндромом.
        Игнорирует регистр. Пустая строка считается палиндромом.
        """
        s_lower = self.lower()
        return s_lower == s_lower[::-1]
s = SuperStr("123123123123")
print(s.is_repeatance("123")) # True
print(s.is_repeatance("123123")) # True
print(s.is_repeatance("123123123123")) # True
print(s.is_repeatance("12312")) # False
print(s.is_repeatance(123)) # False
print(s.is_palindrom()) # False
print(s) # 123123123123 (строка)
print(int(s)) # 123123123123 (целое число)
print(s + "qwe") # 123123123123qwe
p = SuperStr("123_321")
print(p.is_palindrom()) # True
````

### **3.1. Наследование**

* Класс Mentor должен стать родительским классом для преподавателей. От него нужно реализовать наследование классов Lecturer (лекторы) и Reviewer (эксперты, проверяющие домашние задания).

> Имя, фамилю и список закрепленных курсов логично реализовать на уровне родительского класса.

````python
class Student:
    def __init__(self, name, surname, gender):
        self.name = name
        self.surname = surname
        self.gender = gender
        self.finished_courses = []
        self.courses_in_progress = []
        self.grades = {}  # {курс: [оценки]}

    def add_courses(self, course_name):
        """Добавляет курс в список завершённых."""
        self.finished_courses.append(course_name)

    def rate_lecturer(self, lecturer, course, grade):
        """
        Студент может поставить оценку лектору,
        если лектор действительно прикреплён к этому курсу.
        """
        if isinstance(lecturer, Lecturer) and course in lecturer.courses_attached and course in self.courses_in_progress:
            lecturer.grades.setdefault(course, []).append(grade)
        else:
            print("Ошибка при выставлении оценки лектору.")

    def __str__(self):
        res = (f"Имя: {self.name}\n"
               f"Фамилия: {self.surname}\n"
               f"Курсы в процессе изучения: {', '.join(self.courses_in_progress)}\n"
               f"Завершённые курсы: {', '.join(self.finished_courses)}")
        return res


class Mentor:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
        self.courses_attached = []  # список курсов, с которыми работает преподаватель


class Lecturer(Mentor):
    def __init__(self, name, surname):
        super().__init__(name, surname)
        self.grades = {}  # {курс: [оценки студентов]}

    def average_grade(self):
        """Вычисляет среднюю оценку лектора по всем курсам."""
        all_grades = [grade for grades in self.grades.values() for grade in grades]
        return round(sum(all_grades) / len(all_grades), 2) if all_grades else 0

    def __str__(self):
        res = (f"Имя: {self.name}\n"
               f"Фамилия: {self.surname}\n"
               f"Средняя оценка за лекции: {self.average_grade()}")
        return res


class Reviewer(Mentor):
    def rate_hw(self, student, course, grade):
        """
        Проверяющий ставит оценку студенту по курсу, который у него закреплён.
        """
        if isinstance(student, Student) and course in self.courses_attached and course in student.courses_in_progress:
            student.grades.setdefault(course, []).append(grade)
        else:
            print("Ошибка при выставлении оценки студенту.")

    def __str__(self):
        return f"Имя: {self.name}\nФамилия: {self.surname}"


# === Пример использования ===

# создаём студентов
student_1 = Student("Иван", "Иванов", "м")
student_1.courses_in_progress += ["Python"]
student_1.finished_courses += ["Git"]

student_2 = Student("Мария", "Петрова", "ж")
student_2.courses_in_progress += ["Python"]

# создаём преподавателей
lecturer_1 = Lecturer("Сергей", "Орлов")
lecturer_1.courses_attached += ["Python"]

reviewer_1 = Reviewer("Анна", "Смирнова")
reviewer_1.courses_attached += ["Python"]

# эксперты ставят оценки студентам
reviewer_1.rate_hw(student_1, "Python", 9)
reviewer_1.rate_hw(student_2, "Python", 8)

# студенты оценивают лектора
student_1.rate_lecturer(lecturer_1, "Python", 10)
student_2.rate_lecturer(lecturer_1, "Python", 9)

# вывод информации
print("=== Проверяющий ===")
print(reviewer_1)
print("\n=== Лектор ===")
print(lecturer_1)
print("\n=== Студенты ===")
print(student_1)
print()
print(student_2)
````

### **3.2. Атрибуты и взаимодействие классов**

* Реализуйте возможность проставления оценок студентам только Reviewer-ам (реализуйте такой метод).

* Реализуйте метод выставления оценок лекторам у класса Student (оценки по 10-балльной шкале, хранятся в атрибуте-словаре у Lecturer, в котором ключи – названия курсов, а значения – списки оценок). Лектор при этом должен быть закреплен за тем курсом, на который записан студент (реализовать проверку).

````python
from typing import List, Dict, Tuple
import math


class Student:
    def __init__(self, name: str, surname: str, gender: str):
        self.name: str = name
        self.surname: str = surname
        self.gender: str = gender
        self.finished_courses: List[str] = []
        self.courses_in_progress: List[str] = []
        self.grades: Dict[str, List[float]] = {}  # оценки студента по курсам

    def add_courses(self, course_name: str) -> None:
        """Добавляет курс в список завершённых."""
        self.finished_courses.append(course_name)

    def rate_lecturer(self, lecturer: "Lecturer", course: str, grade: float) -> None:
        """
        Студент ставит оценку лектору.
        Проверки:
          - lecturer должен быть экземпляром Lecturer
          - курс должен быть закреплён за лектором
          - курс должен быть в courses_in_progress у студента
          - оценка должна быть в диапазоне 0..10
        """
        if not isinstance(lecturer, Lecturer):
            print("Ошибка: можно оценивать только лекторов.")
            return

        if course not in self.courses_in_progress:
            print(f"Ошибка: вы не проходите курс '{course}', оценка не поставлена.")
            return

        if course not in lecturer.courses_attached:
            print(f"Ошибка: лектор не ведёт курс '{course}', оценка не поставлена.")
            return

        if not (0 <= grade <= 10):
            print("Ошибка: оценка должна быть в диапазоне 0..10.")
            return

        lecturer.grades.setdefault(course, []).append(float(grade))

    def average_grade(self) -> float:
        """Средняя оценка студента по всем курсам."""
        all_grades = [g for grades in self.grades.values() for g in grades]
        return round(sum(all_grades) / len(all_grades), 2) if all_grades else 0.0

    def __str__(self) -> str:
        return (f"Студент: {self.name} {self.surname}\n"
                f"Курсы в процессе: {', '.join(self.courses_in_progress)}\n"
                f"Завершённые курсы: {', '.join(self.finished_courses)}\n"
                f"Средняя оценка: {self.average_grade()}")


class Mentor:
    def __init__(self, name: str, surname: str):
        self.name: str = name
        self.surname: str = surname
        self.courses_attached: List[str] = []  # курсы, за которые закреплён ментор

    def __str__(self) -> str:
        return f"{self.__class__.__name__}: {self.name} {self.surname}"


class Lecturer(Mentor):
    def __init__(self, name: str, surname: str):
        super().__init__(name, surname)
        self.grades: Dict[str, List[float]] = {}  # оценки, поставленные студентами {курс: [оценки]}

    def average_grade(self) -> float:
        """Средняя оценка лектора по всем курсам."""
        all_grades = [g for grades in self.grades.values() for g in grades]
        return round(sum(all_grades) / len(all_grades), 2) if all_grades else 0.0

    def __str__(self) -> str:
        return (f"Лектор: {self.name} {self.surname}\n"
                f"Курсы: {', '.join(self.courses_attached)}\n"
                f"Средняя оценка за лекции: {self.average_grade()}")


class Reviewer(Mentor):
    def rate_hw(self, student: Student, course: str, grade: float) -> None:
        """
        Ревьюер ставит оценку студенту.
        Проверки:
          - student должен быть экземпляром Student
          - курс должен быть закреплён за ревьюером
          - курс должен быть в courses_in_progress у студента
          - оценка должна быть в диапазоне 0..10
        """
        if not isinstance(student, Student):
            print("Ошибка: можно оценивать только студентов.")
            return

        if course not in self.courses_attached:
            print(f"Ошибка: вы не прикреплены к курсу '{course}', оценка не поставлена.")
            return

        if course not in student.courses_in_progress:
            print(f"Ошибка: студент не проходит курс '{course}', оценка не поставлена.")
            return

        if not (0 <= grade <= 10):
            print("Ошибка: оценка должна быть в диапазоне 0..10.")
            return

        student.grades.setdefault(course, []).append(float(grade))

    def __str__(self) -> str:
        return f"Ревьюер: {self.name} {self.surname}\nКурсы: {', '.join(self.courses_attached)}"


# === Демонстрация использования ===
if __name__ == "__main__":
    # создаём студентов
    s1 = Student("Иван", "Иванов", "м")
    s1.courses_in_progress.append("Python")
    s1.finished_courses.append("Git")

    s2 = Student("Мария", "Петрова", "ж")
    s2.courses_in_progress.append("Python")

    # создаём лекторов и ревьюера
    lec = Lecturer("Сергей", "Орлов")
    lec.courses_attached.append("Python")

    rev = Reviewer("Анна", "Смирнова")
    rev.courses_attached.append("Python")

    # корректные действия
    rev.rate_hw(s1, "Python", 9)           # ревьюер ставит студенту оценку
    rev.rate_hw(s2, "Python", 8.5)

    s1.rate_lecturer(lec, "Python", 10)    # студент ставит лектору оценку
    s2.rate_lecturer(lec, "Python", 9.5)

    # некорректные попытки
    rev.rate_hw(s1, "Math", 7)             # ревьюер не прикреплён к курсу Math
    rev.rate_hw("not a student", "Python", 7)  # не студент
    rev.rate_hw(s1, "Python", 15)          # оценка вне диапазона

    s1.rate_lecturer(lec, "Math", 8)       # студент не проходит Math
    s1.rate_lecturer("not a lecturer", "Python", 8)  # не лектор
    s1.rate_lecturer(lec, "Python", -1)    # оценка вне диапазона

    # выводим данные
    print()
    print(rev)
    print()
    print(lec)
    print()
    print(s1)
    print()
    print(s2)

    # смотрим внутренние структуры оценок
    print("\nОценки студентов (s1):", s1.grades)
    print("Оценки лекторов (lec):", lec.grades)
````

Student.__len__ — удобно узнать, сколько оценок у студента (используется len(student)).

Student.__lt__, Student.__eq__, Lecturer.__lt__, Lecturer.__eq__ — логичная семантика: сравнение по средней оценке; поддерживается сравнение между объектами, имеющими average_grade() (например, между студентом и лектором).

````python
from typing import List, Dict, Optional


class Student:
    def __init__(self, name: str, surname: str, gender: str):
        self.name: str = name
        self.surname: str = surname
        self.gender: str = gender
        self.finished_courses: List[str] = []
        self.courses_in_progress: List[str] = []
        self.grades: Dict[str, List[float]] = {}  # {course: [grades]}

    def add_courses(self, course_name: str) -> None:
        self.finished_courses.append(course_name)

    def rate_lecturer(self, lecturer: "Lecturer", course: str, grade: float) -> None:
        if not isinstance(lecturer, Lecturer):
            print("Ошибка: можно оценивать только лекторов.")
            return
        if course not in self.courses_in_progress:
            print(f"Ошибка: студент не проходит курс '{course}', оценка не поставлена.")
            return
        if course not in lecturer.courses_attached:
            print(f"Ошибка: лектор не ведёт курс '{course}', оценка не поставлена.")
            return
        if not (0 <= grade <= 10):
            print("Ошибка: оценка должна быть в диапазоне 0–10.")
            return

        lecturer.grades.setdefault(course, []).append(float(grade))

    def average_grade(self) -> float:
        all_grades = [g for grades in self.grades.values() for g in grades]
        return round(sum(all_grades) / len(all_grades), 2) if all_grades else 0.0

    def __len__(self) -> int:
        return sum(len(lst) for lst in self.grades.values())

    def __lt__(self, other) -> bool:
        if hasattr(other, "average_grade"):
            return self.average_grade() < other.average_grade()
        return NotImplemented

    def __eq__(self, other) -> bool:
        if hasattr(other, "average_grade"):
            return self.average_grade() == other.average_grade()
        return NotImplemented

    def __str__(self) -> str:
        avg = self.average_grade()
        in_progress = ", ".join(self.courses_in_progress) or "Нет"
        finished = ", ".join(self.finished_courses) or "Нет"
        return (f"Имя: {self.name}\n"
                f"Фамилия: {self.surname}\n"
                f"Средняя оценка за домашние задания: {avg}\n"
                f"Курсы в процессе изучения: {in_progress}\n"
                f"Завершенные курсы: {finished}")


class Mentor:
    def __init__(self, name: str, surname: str):
        self.name: str = name
        self.surname: str = surname
        self.courses_attached: List[str] = []

    def __contains__(self, course: str) -> bool:
        return course in self.courses_attached

    def __str__(self) -> str:
        return f"Имя: {self.name}\nФамилия: {self.surname}"


class Lecturer(Mentor):
    def __init__(self, name: str, surname: str):
        super().__init__(name, surname)
        self.grades: Dict[str, List[float]] = {}

    def average_grade(self) -> float:
        all_grades = [g for grades in self.grades.values() for g in grades]
        return round(sum(all_grades) / len(all_grades), 2) if all_grades else 0.0

    def __lt__(self, other) -> bool:
        if hasattr(other, "average_grade"):
            return self.average_grade() < other.average_grade()
        return NotImplemented

    def __eq__(self, other) -> bool:
        if hasattr(other, "average_grade"):
            return self.average_grade() == other.average_grade()
        return NotImplemented

    def __str__(self) -> str:
        avg = self.average_grade()
        return (f"Имя: {self.name}\n"
                f"Фамилия: {self.surname}\n"
                f"Средняя оценка за лекции: {avg}")


class Reviewer(Mentor):
    def rate_hw(self, student: Student, course: str, grade: float) -> None:
        if not isinstance(student, Student):
            print("Ошибка: можно оценивать только студентов.")
            return
        if course not in self.courses_attached:
            print(f"Ошибка: ревьюер не ведёт курс '{course}', оценка не поставлена.")
            return
        if course not in student.courses_in_progress:
            print(f"Ошибка: студент не проходит курс '{course}', оценка не поставлена.")
            return
        if not (0 <= grade <= 10):
            print("Ошибка: оценка должна быть в диапазоне 0–10.")
            return

        student.grades.setdefault(course, []).append(float(grade))

    def __str__(self) -> str:
        return f"Имя: {self.name}\nФамилия: {self.surname}"


# --- Функции для средней оценки по курсу ---

def avg_student_grade_for_course(students: List[Student], course: str) -> Optional[float]:
    """
    Средняя оценка за домашние задания по курсу среди переданных студентов.
    Возвращает float (округлённый до 2 знаков) или None, если оценок нет.
    """
    collected: List[float] = []
    for s in students:
        grades = s.grades.get(course, [])
        collected.extend(grades)
    if not collected:
        return None
    return round(sum(collected) / len(collected), 2)


def avg_lecturer_grade_for_course(lecturers: List[Lecturer], course: str) -> Optional[float]:
    """
    Средняя оценка за лекции по курсу среди переданных лекторов.
    Возвращает float или None, если оценок нет.
    """
    collected: List[float] = []
    for l in lecturers:
        grades = l.grades.get(course, [])
        collected.extend(grades)
    if not collected:
        return None
    return round(sum(collected) / len(collected), 2)


# === Демонстрация: создаём по 2 экземпляра каждого класса и вызываем все методы ===

if __name__ == "__main__":
    # Студенты
    s1 = Student("Руой", "Эман", "м")
    s1.courses_in_progress += ["Python", "Git"]
    s1.finished_courses += ["Введение в программирование"]

    s2 = Student("Иван", "Иванов", "м")
    s2.courses_in_progress += ["Python"]
    s2.finished_courses += ["Основы алгоритмов"]

    # Лекторы
    l1 = Lecturer("Сергей", "Орлов")
    l1.courses_attached += ["Python"]

    l2 = Lecturer("Ольга", "Кузнецова")
    l2.courses_attached += ["Git", "Python"]  # 2 курса

    # Ревьюеры
    r1 = Reviewer("Анна", "Смирнова")
    r1.courses_attached += ["Python"]

    r2 = Reviewer("Владимир", "Сидоров")
    r2.courses_attached += ["Git"]

    # Выставляем оценки студентам (только ревьюеры)
    r1.rate_hw(s1, "Python", 9)
    r1.rate_hw(s1, "Python", 10)
    r1.rate_hw(s2, "Python", 8.5)

    r2.rate_hw(s1, "Git", 8)
    # Попытка ревьюера ставить оценку по курсу, к которому он не прикреплён
    r2.rate_hw(s2, "Git", 7)  # s2 не проходит Git -> сообщение об ошибке

    # Студенты оценивают лекторов (только студенты)
    s1.rate_lecturer(l1, "Python", 10)
    s1.rate_lecturer(l2, "Git", 9)
    s2.rate_lecturer(l1, "Python", 9)
    # Некорректная попытка: s2 пытается оценить l2 по Git, но s2 не проходит Git
    s2.rate_lecturer(l2, "Git", 8)  # ожидаем сообщение об ошибке

    # Печать информации через __str__
    print("=== Проверяющие ===")
    print(r1)
    print()
    print(r2)
    print("\n=== Лекторы ===")
    print(l1)
    print()
    print(l2)
    print("\n=== Студенты ===")
    print(s1)
    print()
    print(s2)

    # Демонстрация магических методов
    print(f"\nКоличество оценок у {s1.name}: {len(s1)}")  # __len__
    print(f"{s1.name} > {s2.name}? -> {s1 > s2}")         # __lt__/__gt__ via comparisons
    print(f"{l1.name} == {l2.name}? -> {l1 == l2}")      # сравнение лекторов по среднему

    # Проверка contains для ментора
    print(f"\n'Python' in l1? -> {'Python' in l1}")
    print(f"'Git' in l1? -> {'Git' in l1}")

    # Внутренние структуры оценок
    print("\nОценки студентов (s1):", s1.grades)
    print("Оценки студентов (s2):", s2.grades)
    print("Оценки лекторов (l1):", l1.grades)
    print("Оценки лекторов (l2):", l2.grades)

    # --- использование функций подсчёта средней по курсу ---
    students = [s1, s2]
    lecturers = [l1, l2]

    course_name = "Python"
    avg_stud_python = avg_student_grade_for_course(students, course_name)
    avg_lec_python = avg_lecturer_grade_for_course(lecturers, course_name)

    print(f"\nСредняя оценка студентов за домашние задания по курсу '{course_name}':",
          avg_stud_python if avg_stud_python is not None else "Оценок нет")

    print(f"Средняя оценка лекторов за лекции по курсу '{course_name}':",
          avg_lec_python if avg_lec_python is not None else "Оценок нет")

    course_name = "Git"
    avg_stud_git = avg_student_grade_for_course(students, course_name)
    avg_lec_git = avg_lecturer_grade_for_course(lecturers, course_name)

    print(f"\nСредняя оценка студентов за домашние задания по курсу '{course_name}':",
          avg_stud_git if avg_stud_git is not None else "Оценок нет")

    print(f"Средняя оценка лекторов за лекции по курсу '{course_name}':",
          avg_lec_git if avg_lec_git is not None else "Оценок нет")
````

````python
from __future__ import annotations
from typing import Union
import math
from fractions import Fraction as _PyFraction  # для точного преобразования float -> дробь


Number = Union[int, float, "Fraction"]


def _gcd(a: int, b: int) -> int:
    """Наибольший общий делитель (положительный)."""
    return math.gcd(a, b)


class Fraction:
    """
    Класс для работы с обыкновенными дробями.
    Хранит несократимую дробь numerator/denominator, denominator > 0.
    """

    def __init__(self, numerator: int, denominator: int = 1):
        if not isinstance(numerator, int) or not isinstance(denominator, int):
            raise TypeError("Числитель и знаменатель должны быть целыми числами.")
        if denominator == 0:
            # Требование: при попытке создать дробь с нулевым знаменателем выводить в консоль сообщение ValueError.
            raise ValueError("Знаменатель не может быть нулем")
        # нормализация знака: знаменатель положительный
        if denominator < 0:
            numerator = -numerator
            denominator = -denominator
        # сокращаем дробь
        g = _gcd(abs(numerator), denominator)
        self.numerator = numerator // g
        self.denominator = denominator // g

    # ---- вспомогательные ----
    @classmethod
    def _ensure_fraction(cls, value: Number) -> "Fraction":
        """Возвращает Fraction: если value уже Fraction — возвращает как есть;
           если int/float — конвертирует.
           Для float используем точное приближение через fractions.Fraction.
        """
        if isinstance(value, Fraction):
            return value
        if isinstance(value, int):
            return Fraction(value, 1)
        if isinstance(value, float):
            pyf = _PyFraction(value).limit_denominator()
            return Fraction(pyf.numerator, pyf.denominator)
        raise TypeError("Операция поддерживает только Fraction, int или float.")

    def _binary_op(self, other: Number, op: str) -> "Fraction":
        other_f = self._ensure_fraction(other)
        a, b = self.numerator, self.denominator
        c, d = other_f.numerator, other_f.denominator

        if op == "add":
            num = a * d + b * c
            den = b * d
        elif op == "sub":
            num = a * d - b * c
            den = b * d
        elif op == "mul":
            num = a * c
            den = b * d
        elif op == "truediv":
            if c == 0:
                # Деление на нулевую дробь — сообщение ZeroDivisionError.
                raise ZeroDivisionError("Деление на нулевую дробь")
            num = a * d
            den = b * c
        else:
            raise ValueError("Неизвестная операция")
        return Fraction(num, den)

    # ---- арифметика ----
    def __add__(self, other: Number) -> "Fraction":
        return self._binary_op(other, "add")

    def __radd__(self, other: Number) -> "Fraction":
        return self._ensure_fraction(other).__add__(self)

    def __sub__(self, other: Number) -> "Fraction":
        return self._binary_op(other, "sub")

    def __rsub__(self, other: Number) -> "Fraction":
        return self._ensure_fraction(other).__sub__(self)

    def __mul__(self, other: Number) -> "Fraction":
        return self._binary_op(other, "mul")

    def __rmul__(self, other: Number) -> "Fraction":
        return self._ensure_fraction(other).__mul__(self)

    def __truediv__(self, other: Number) -> "Fraction":
        return self._binary_op(other, "truediv")

    def __rtruediv__(self, other: Number) -> "Fraction":
        return self._ensure_fraction(other).__truediv__(self)

    # ---- сравнения ----
    def __eq__(self, other: object) -> bool:
        if not isinstance(other, (Fraction, int, float)):
            return NotImplemented
        other_f = self._ensure_fraction(other)  # type: ignore[arg-type]
        return self.numerator == other_f.numerator and self.denominator == other_f.denominator

    def __lt__(self, other: Number) -> bool:
        other_f = self._ensure_fraction(other)
        return self.numerator * other_f.denominator < other_f.numerator * self.denominator

    def __le__(self, other: Number) -> bool:
        other_f = self._ensure_fraction(other)
        return self.numerator * other_f.denominator <= other_f.numerator * self.denominator

    def __gt__(self, other: Number) -> bool:
        other_f = self._ensure_fraction(other)
        return self.numerator * other_f.denominator > other_f.numerator * self.denominator

    def __ge__(self, other: Number) -> bool:
        other_f = self._ensure_fraction(other)
        return self.numerator * other_f.denominator >= other_f.numerator * self.denominator

    # ---- преобразования ----
    def __float__(self) -> float:
        return self.numerator / self.denominator

    @classmethod
    def from_float(cls, value: float) -> "Fraction":
        """Построение Fraction из float (используем точное приближение stdlib)."""
        if not isinstance(value, float):
            raise TypeError("from_float принимает float")
        pyf = _PyFraction(value).limit_denominator()
        return cls(pyf.numerator, pyf.denominator)

    def reciprocal(self) -> "Fraction":
        """Возвращает новую дробь, обратную данной."""
        if self.numerator == 0:
            raise ZeroDivisionError("Обратная дробь не существует для нулевой дроби")
        return Fraction(self.denominator, self.numerator)

    # ---- строковые представления ----
    def __str__(self) -> str:
        if self.denominator == 1:
            return f"{self.numerator}"
        return f"{self.numerator}/{self.denominator}"

    def __repr__(self) -> str:
        return f"Fraction({self.numerator}, {self.denominator})"
if __name__ == "__main__":
    print("=== Создание дробей ===")
    a = Fraction(3, 4)
    b = Fraction(2, 3)
    c = Fraction(-6, 8)  # автоматически сокращается до -3/4
    print("a =", a)   # 3/4
    print("b =", b)   # 2/3
    print("c =", c)   # -3/4

    print("\n=== Арифметика ===")
    print("a + b =", a + b)        # (3/4 + 2/3) = 17/12
    print("a - b =", a - b)        # (3/4 - 2/3) = 1/12
    print("a * b =", a * b)        # 1/2
    print("a / b =", a / b)        # 9/8

    print("\n=== Сравнения ===")
    print("a == c?", a == c)       # False
    print("a > b?", a > b)         # 3/4 > 2/3 -> True
    print("c < 0?", c < 0)         # True
    print("a >= 0.75?", a >= 0.75) # True (0.75 == 3/4)

    print("\n=== Работа с числами int/float ===")
    print("a + 1 =", a + 1)        # 7/4
    print("2 + b =", 2 + b)        # 8/3
    print("a * 2 =", a * 2)        # 3/2
    print("float(a) =", float(a))  # 0.75

    print("\n=== Reciprocal и from_float ===")
    print("reciprocal(a) =", a.reciprocal())  # 4/3
    f = Fraction.from_float(0.125)            # 1/8
    print("from_float(0.125) =", f)

    print("\n=== __repr__ ===")
    print(repr(a))  # Fraction(3, 4)

    print("\n=== Обработка исключений при создании и при делении ===")
    try:
        bad = Fraction(1, 0)
    except ValueError as e:
        print("Ошибка при создании дроби:", e)

    zero = Fraction(0, 1)
    try:
        print("a / zero =", a / zero)
    except ZeroDivisionError as e:
        print("Ошибка при делении:", e)

    # деление 0/1 reciprocal
    try:
        print("reciprocal(zero) =", zero.reciprocal())
    except ZeroDivisionError as e:
        print("Ошибка при взятии обратной:", e)

    print("\n=== Дополнительные проверки ===")
    # Сокращение:
    x = Fraction(100, 250)  # 2/5
    print("100/250 ->", x)
    # Знак в числителе:
    y = Fraction(1, -3)
    print("1/-3 ->", y)  # -1/3
````
