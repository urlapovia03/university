В качестве входных данных для разрабатываемой методики автоматизированного транспортного районирования используются разнородные пространственные, транспортные и социально-экономические данные, обеспечивающие комплексное описание городской территории.

### Основные типы входных данных:

#### 1. Геопространственные данные:
- улично-дорожная сеть (граф транспортной сети);

#### 3. Социально-экономические данные:

#### 4. Данные мобильности и трекинга (при наличии):
